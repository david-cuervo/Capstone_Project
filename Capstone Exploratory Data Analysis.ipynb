## Abstract

### Objective
Explore the relationships between income, race, and industrial toxic releases in the state of Texas.
### Methods
 - Explore carcinogen release

 - Plot the polluting parent companys
 - Plot the most polluted counties
 - Explore Harris county (most polluted county)
 - Explore the Census data
 - Merge Census and EPA data

### Conclusions
From the analysis conducted in the Modeling notebook, the results were inconclusive. The best model had an accuracy of 51% and the important features in the model were unclear. From this data, it is unclear if industrial toxic releases are more prevelant in poorer ZIP codes or ZIP codes with a higher population of people of color. Future analyses can study other sources of pollution such as municipal waste sites, bus depots, and water treatment facilities.
